## Image and its origin coordinate frame

```python
origin = faai.se2([0, 0], 0)

origin
```

```python
plt.imshow(im)
plt.axis('off')

faai.plot_frame(origin, scale_factor=100, c='c')

plt.show()
```

## Origin manipulation in `plt.imshow`

```python
spoon_pose = faai.se2([410, 75], np.pi/12)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.imshow(im)

faai.plot_frame(spoon_pose, scale_factor=100, c='b')

plt.subplot(1, 2, 2)
plt.imshow(im, origin='lower')

faai.plot_frame(spoon_pose, scale_factor=100, c='b')

plt.show()
```
