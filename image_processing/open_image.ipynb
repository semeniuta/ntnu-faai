## Show image channels

```python
def show_channels(im, colormap='gray'):

    plt.figure(figsize=(15, 4))

    plt.subplot(1, 3, 1)
    plt.imshow(im[:, :, 0], cmap=colormap, vmin=0, vmax=255)
    plt.axis('off')
    plt.title('Red')

    plt.subplot(1, 3, 2)
    plt.imshow(im[:, :, 1], cmap=colormap, vmin=0, vmax=255)
    plt.axis('off')
    plt.title('Green')

    plt.subplot(1, 3, 3)
    plt.imshow(im[:, :, 2], cmap=colormap, vmin=0, vmax=255)
    plt.axis('off')
    plt.title('Blue')
```

```python
show_channels(im)
```
