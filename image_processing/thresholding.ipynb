```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from skimage import data
from skimage.filters import threshold_otsu

import faai
from visionfuncs.improc import dilate
from visionfuncs.regions import threshold_binary, threshold_binary_inv
```

## Open image (Scikit-image built-in)

```python
im = data.coins()

print(im.shape)
```

## Show image 

```python
plt.imshow(im, cmap='gray')
plt.show()
```

## Image histogram

```python
im_flattened = im.ravel()
```

```python
plt.hist(im_flattened, bins=50)
plt.show()
```

## Thresholding

```python
threshold_value = 150

im > threshold_value
```

```python
plt.imshow(im > threshold_value)
plt.show()
```

```python
im_t = threshold_binary(im, threshold_value)

plt.imshow(im_t)
plt.show()
```

## Dilation

```python
im_dilated = dilate(im_t, kernel_size=5)

plt.imshow(im_dilated)
plt.show()
```
