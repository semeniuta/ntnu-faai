```python
import numpy as np
from scipy.integrate import solve_ivp
from matplotlib import pyplot as plt
```

## Simple model of a car

p. 234, fig. 6.3

```python
def create_car_diff_eq(m, k, F):

    def equation(t, y):

        x, v = y

        deriv_x = v
        deriv_v = (F - k * v) / m
        
        return np.array([deriv_x, deriv_v])
        
    return equation
```

```python
eq = create_car_diff_eq(m=1000, k=50, F=500)
```

```python
sol = solve_ivp(eq, [0, 120], [0, 0])
```

```python
plt.figure()
plt.plot(sol.t, sol.y[1], 'o-')
plt.title('Velocity')

plt.figure()
plt.plot(sol.t, sol.y[0], 'o-')
plt.title('Position')

plt.show()
```
