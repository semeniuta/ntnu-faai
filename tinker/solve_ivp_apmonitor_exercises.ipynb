```python
import numpy as np
from scipy.integrate import solve_ivp
from matplotlib import pyplot as plt
```

## Problem 1

```python
def model_1(t, y):
    return -y + 1
```

```python
sol_1 = solve_ivp(model_1, t_span=(0, 10), y0=[0])
```

```python
plt.plot(sol_1.t, sol_1.y[0], 'o-')
plt.show()
```

## Problem 2

```python
def model_2(t, y):
    
    coef = 1 / 5.
    
    u = 2. if t >= 10 else 0.
    
    return coef * (-y + u)
```

```python
sol_2 = solve_ivp(model_2, t_span=(0, 50), y0=[1])
```

```python
plt.plot(sol_2.t, sol_2.y[0], 'o-')
plt.show()
```

## Problem 3

```python
def model_3_x(t, x):
    
    return 3. * np.exp(-t)
    

def model_3_y(t, y):
    
    return 3 - y
```

```python
sol_3_x = solve_ivp(model_3_x, t_span=(0, 10), y0=[0])
sol_3_y = solve_ivp(model_3_y, t_span=(0, 10), y0=[0])

plt.plot(sol_3_x.t, sol_3_x.y[0], 'o-')
plt.plot(sol_3_y.t, sol_3_y.y[0], 'o-')
plt.show()
```
