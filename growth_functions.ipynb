```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def exp_decay_increase(upper_limit, k):
    
    def f(t):
        return upper_limit * (1 - np.exp(-k * t))
    
    return f


def logistic_growth(a, b, k):
    
    def f(t):
        return a / (1 + b * np.exp(-k * t))
    
    return f
```

```python
xx = np.arange(0, 10, 0.2)

_ = plt.plot(xx, exp_decay_increase(1., 0.5)(xx))
_ = plt.plot(xx, exp_decay_increase(1., 1.)(xx))
```

```python
xx2 = np.arange(-10, 10, 0.2)

_ = plt.plot(xx2, logistic_growth(1., 2., 0.5)(xx2))
```
